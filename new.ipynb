```python
import os
import shutil
import random
from sklearn.model_selection import train_test_split

# Original dataset path
dataset_path = r"F:\dataset"

# Output split dataset path
output_path = r"F:\dataset_split"

# Train-validation ratio
TRAIN_RATIO = 0.8

# Image extensions
IMG_EXTENSIONS = ('.png', '.jpg', '.jpeg', '.bmp')

# Create train and validation folders
train_dir = os.path.join(output_path, "train")
val_dir = os.path.join(output_path, "validation")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Process each class
for class_name in os.listdir(dataset_path):

    class_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_path):

        # Get all images
        images = [
            f for f in os.listdir(class_path)
            if f.lower().endswith(IMG_EXTENSIONS)
        ]

        # Skip empty classes
        if len(images) == 0:
            continue

        # Split data
        train_imgs, val_imgs = train_test_split(
            images,
            train_size=TRAIN_RATIO,
            random_state=42,
            shuffle=True
        )

        # Create class folders
        os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
        os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

        # Copy train images
        for img in train_imgs:
            src = os.path.join(class_path, img)
            dst = os.path.join(train_dir, class_name, img)
            shutil.copy2(src, dst)

        # Copy validation images
        for img in val_imgs:
            src = os.path.join(class_path, img)
            dst = os.path.join(val_dir, class_name, img)
            shutil.copy2(src, dst)

        print(f"{class_name} --> Train: {len(train_imgs)}, Validation: {len(val_imgs)}")

print("\nDataset splitting completed successfully!")
```

```python
import os
import cv2
import numpy as np

# Input dataset
input_dir = r"F:\dataset"

# Output processed dataset
output_dir = r"F:\processed_dataset"

os.makedirs(output_dir, exist_ok=True)

# ===== Processing Functions =====

def gaussian_blur(img):
    return cv2.GaussianBlur(img, (5,5), 0)

def histogram_equalization(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    eq = cv2.equalizeHist(gray)
    return cv2.cvtColor(eq, cv2.COLOR_GRAY2BGR)

def sharpening(img):
    kernel = np.array([[0,-1,0],
                       [-1,5,-1],
                       [0,-1,0]])
    return cv2.filter2D(img, -1, kernel)

def canny_edge(img):
    edges = cv2.Canny(img, 100, 200)
    return cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)

def morphological(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    kernel = np.ones((3,3), np.uint8)
    morph = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
    return cv2.cvtColor(morph, cv2.COLOR_GRAY2BGR)

processing_methods = {
    "blur": gaussian_blur,
    "hist_eq": histogram_equalization,
    "sharp": sharpening,
    "canny": canny_edge,
    "morph": morphological
}

# ===== Apply Processing =====

for class_name in os.listdir(input_dir):

    class_input = os.path.join(input_dir, class_name)

    if not os.path.isdir(class_input):
        continue

    class_output = os.path.join(output_dir, class_name)
    os.makedirs(class_output, exist_ok=True)

    for image_name in os.listdir(class_input):

        image_path = os.path.join(class_input, image_name)

        img = cv2.imread(image_path)

        if img is None:
            continue

        filename = os.path.splitext(image_name)[0]

        # Apply all processing methods
        for method_name, method in processing_methods.items():

            processed = method(img)

            save_name = f"{filename}_{method_name}.jpg"

            save_path = os.path.join(class_output, save_name)

            cv2.imwrite(save_path, processed)

print("Processing completed!")
```

```python
import os
import cv2
import numpy as np
import random

input_dir = r"F:\processed_dataset"
output_dir = r"F:\augmented_dataset"

os.makedirs(output_dir, exist_ok=True)

# ===== Augmentation Functions =====

def rotate(img):
    angle = random.randint(-15, 15)
    h, w = img.shape[:2]

    M = cv2.getRotationMatrix2D((w//2, h//2), angle, 1)

    return cv2.warpAffine(img, M, (w, h))

def translate(img):
    h, w = img.shape[:2]

    tx = random.randint(-10, 10)
    ty = random.randint(-10, 10)

    M = np.float32([[1,0,tx],[0,1,ty]])

    return cv2.warpAffine(img, M, (w,h))

def gaussian_noise(img):

    noise = np.random.normal(0, 15, img.shape).astype(np.uint8)

    noisy = cv2.add(img, noise)

    return noisy

def sharpness(img):

    kernel = np.array([[0,-1,0],
                       [-1,5,-1],
                       [0,-1,0]])

    return cv2.filter2D(img, -1, kernel)

augmentations = {
    "rot": rotate,
    "trans": translate,
    "noise": gaussian_noise,
    "sharp": sharpness
}

# ===== Apply Augmentations =====

for class_name in os.listdir(input_dir):

    class_input = os.path.join(input_dir, class_name)

    if not os.path.isdir(class_input):
        continue

    class_output = os.path.join(output_dir, class_name)

    os.makedirs(class_output, exist_ok=True)

    for image_name in os.listdir(class_input):

        image_path = os.path.join(class_input, image_name)

        img = cv2.imread(image_path)

        if img is None:
            continue

        filename = os.path.splitext(image_name)[0]

        # Save original processed image
        cv2.imwrite(
            os.path.join(class_output, image_name),
            img
        )

        # Apply augmentations
        for aug_name, aug_func in augmentations.items():

            aug_img = aug_func(img)

            save_name = f"{filename}_{aug_name}.jpg"

            save_path = os.path.join(class_output, save_name)

            cv2.imwrite(save_path, aug_img)

print("Augmentation completed!")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

dataset_path = r"F:\augmented_dataset"
output_path = r"F:\final_dataset"

train_dir = os.path.join(output_path, "train")
val_dir = os.path.join(output_path, "validation")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

IMG_EXTENSIONS = ('.png', '.jpg', '.jpeg', '.bmp')

for class_name in os.listdir(dataset_path):

    class_path = os.path.join(dataset_path, class_name)

    if not os.path.isdir(class_path):
        continue

    images = [
        f for f in os.listdir(class_path)
        if f.lower().endswith(IMG_EXTENSIONS)
    ]

    train_imgs, val_imgs = train_test_split(
        images,
        test_size=0.2,
        random_state=42,
        shuffle=True
    )

    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

    for img in train_imgs:
        shutil.copy2(
            os.path.join(class_path, img),
            os.path.join(train_dir, class_name, img)
        )

    for img in val_imgs:
        shutil.copy2(
            os.path.join(class_path, img),
            os.path.join(val_dir, class_name, img)
        )

print("Dataset split completed!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader, Subset

# ===== Device =====
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("Using Device:", device)

# ===== Paths =====
train_dir = r"F:\final_dataset\train"
val_dir = r"F:\final_dataset\validation"

# ===== FAST SETTINGS =====
IMG_SIZE = 64
BATCH_SIZE = 64
EPOCHS = 3
LEARNING_RATE = 0.001

# ===== Transform =====
transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

# ===== Dataset =====
train_dataset = datasets.ImageFolder(
    train_dir,
    transform=transform
)

val_dataset = datasets.ImageFolder(
    val_dir,
    transform=transform
)

# ===== OPTIONAL: USE SMALLER SUBSET =====
# Use only part of dataset for faster training

MAX_TRAIN = 5000
MAX_VAL = 1000

train_dataset = Subset(
    train_dataset,
    range(min(MAX_TRAIN, len(train_dataset)))
)

val_dataset = Subset(
    val_dataset,
    range(min(MAX_VAL, len(val_dataset)))
)

# ===== DataLoader =====
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=0
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=0
)

# ===== MobileNetV2 =====
model = models.mobilenet_v2(weights="DEFAULT")

# Freeze all layers
for param in model.features.parameters():
    param.requires_grad = False

# Replace classifier
num_classes = len(
    datasets.ImageFolder(train_dir).classes
)

model.classifier = nn.Sequential(
    nn.Linear(model.last_channel, num_classes)
)

model = model.to(device)

# ===== Loss =====
criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=LEARNING_RATE
)

# ===== TRAIN =====
for epoch in range(EPOCHS):

    model.train()

    correct = 0
    total = 0

    for images, labels in train_loader:

        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

    train_acc = 100 * correct / total

    print(f"Epoch {epoch+1}/{EPOCHS}")
    print(f"Train Accuracy: {train_acc:.2f}%")

# ===== VALIDATION =====
model.eval()

correct = 0
total = 0

with torch.no_grad():

    for images, labels in val_loader:

        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)

        _, predicted = torch.max(outputs, 1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

val_acc = 100 * correct / total

print(f"\nValidation Accuracy: {val_acc:.2f}%")

# ===== Save =====
torch.save(model.state_dict(), "fast_mobilenet.pth")

print("Training Completed!")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

IMG_SIZE = 64
BATCH_SIZE = 64

transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

train_dir = r"F:\final_dataset\train"
val_dir = r"F:\final_dataset\validation"

train_dataset = datasets.ImageFolder(train_dir, transform=transform)
val_dataset = datasets.ImageFolder(val_dir, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

print("DataLoaders recreated successfully")
```

```python
def evaluate(loader):
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)

            _, predicted = torch.max(outputs, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return 100 * correct / total


train_acc = evaluate(train_loader)
val_acc = evaluate(val_loader)

print("===== FINAL ACCURACY =====")
print(f"Train Accuracy: {train_acc:.2f}%")
print(f"Validation Accuracy: {val_acc:.2f}%")
```
